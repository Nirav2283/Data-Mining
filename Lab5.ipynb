### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
puredata = df.copy()
puredata['Age'] = puredata['Age'].interpolate()
puredata
```

```python
##min max scaling
data_minmax = puredata.copy()
age_mm = data_minmax['Age']
age_min = age_mm.min()
age_max = age_mm.max()
data_minmax['MinMaxAge'] = (age_mm - age_min) / (age_max - age_min)
data_minmax[['Age' , 'MinMaxAge']]
```

```python
##decimal scaling
decimal_scaling = puredata.copy()
age_ds = decimal_scaling['Age']
max_age = age_ds.max()
temp = len(str(int(max_age)))
##int is necesaary otherwise it will give 4
print(max_age)
print(temp)
decimal_scaling['DecimalScaleAge'] = age_ds / (10 ** temp)
decimal_scaling[['Age' , 'DecimalScaleAge']]
```
